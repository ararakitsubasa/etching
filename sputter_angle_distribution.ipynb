```python
def emission_angle(theta1, theta, phi):
    Eth = 114
    E = 250
    # phi = np.pi/3
    gamma = (3*np.sin(theta1)**2 - 1)/np.sin(theta1)**2 + \
            (np.cos(theta1)**2) * (3*np.sin(theta1)**2 + 1)/(2*np.sin(theta1)**3)*np.log((1+ np.sin(theta1))/(1-np.sin(theta1)))

    return np.cos(theta)*(1 - (0.25*(Eth/E)**0.5)*(np.cos(theta) * gamma + 1.5*np.pi*np.sin(theta)*np.sin(theta1)*np.cos(phi) )) 

theta = np.arange(-np.pi/2, np.pi/2, 0.1)
y = emission_angle(np.pi/4, theta, 0)


fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, y)
# ax.plot(y, theta)
# plt.plot(theta, y)
ax.quiver(0,0, 1, -1,
        scale=1., zorder=3, color='red', pivot= 'tip',
          width=0.01,)

ax.set_theta_zero_location('N')
```

```python
def sputterYield_Func(gamma0, gammaMax, thetaMax):
    f = -np.log(gammaMax/gamma0)/(np.log(np.cos(gammaMax)) + 1 - np.cos(thetaMax))
    s = f*np.cos(thetaMax)
    theta = np.arange(0, np.pi/2, 0.01)
    sputterYield = gamma0*np.cos(theta)**(-f)*np.exp(-s*(1/np.cos(theta) - 1))
    sputterYield[-1] = 0
    theta[-1] = np.pi/2
    yield_hist = np.zeros((2, theta.shape[0]))

    yield_hist[0, :] = sputterYield
    yield_hist[1, :] = theta
    return yield_hist
```
